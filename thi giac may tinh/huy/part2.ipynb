```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_marketing = pd.read_csv('./Marketing.csv')
data_marketing.head()
```

```python
def write_model(file_name = ''):
    data = data_marketing[[file_name, 'AmountSpent']]
    data = data.groupby(file_name).sum()

    # plot
    data.plot(kind='bar', figsize=(20, 10))

    plt.xlabel(file_name) 
    plt.ylabel('AmountSpent')

    plt.title('AmountSpent by ',file_name)

    plt.show()

```

```python
write_model('Age')
```
