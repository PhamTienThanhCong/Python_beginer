```python
# read file tomato-yield.csv by using pandas
# and print the first 5 rows of the data

import pandas as pd
```

```python
data_tomato = pd.read_csv('./tomato-yields.csv')
data_tomato.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_tomato_year = data_tomato.groupby('Year').mean()

data_tomato_year.plot(kind='bar', figsize=(10, 5))
plt.xlabel('Year')
plt.ylabel('Yield')
plt.title('Average Tomato Yield by Year')
plt.show()
```

```python
# Summary the data, including basic statistic value and quartile value in data
data_tomato.describe()
```

```python
# The country has largest value of tomato yield in 2000
data_tomato_2000 = data_tomato[data_tomato['Year'] == 2000]
data_tomato_2000.head()
```

```python
result = data_tomato_2000[data_tomato_2000['Tomatoes | 00000388 || Yield | 005419 || tonnes per hectare'] == data_tomato_2000['Tomatoes | 00000388 || Yield | 005419 || tonnes per hectare'].max()]

print(result['Entity'])
```

```python
# The country has the largest total value of tomato yield during 1961-2000

data_tomato_1961_2000 = data_tomato[(data_tomato['Year'] >= 1961) & (data_tomato['Year'] <= 2000)]
data_tomato_1961_2000.head()
```

```python
data_tomato_1961_2000 = data_tomato_1961_2000.groupby('Entity').sum()
data_tomato_1961_2000.head()
```

```python
result = data_tomato_1961_2000[data_tomato_1961_2000['Tomatoes | 00000388 || Yield | 005419 || tonnes per hectare'] == data_tomato_1961_2000['Tomatoes | 00000388 || Yield | 005419 || tonnes per hectare'].max()]

print("Max 1961 to 2000: ",result.index[0])
```

```python
# Obtain a line chart of yeild of several countries in Asia
data_tomato_asia = data_tomato[data_tomato['Entity'].isin(['China', 'India'])]
data_tomato_asia.head()
```

```python

data_tomato_asia = data_tomato_asia.groupby(['Entity', 'Year']).mean()
data_tomato_asia
```

```python
# write a line chart china

data_tomato_asia.loc['China'].plot(kind='line', figsize=(10, 5))
plt.xlabel('Year')
plt.ylabel('Yield')
plt.title('Average Tomato Yield by Year in China')
plt.show()
```

```python
data_tomato_asia.loc['India'].plot(kind='line', figsize=(10, 5))
plt.xlabel('Year')
plt.ylabel('Yield')
plt.title('Average Tomato Yield by Year in India')
plt.show()
```

```python
# write a line chart have two line in one chart
data_tomato_asia.loc['China'].plot(kind='line', figsize=(10, 5))
data_tomato_asia.loc['India'].plot(kind='line', figsize=(10, 5))
plt.xlabel('Year')
plt.ylabel('Yield')
plt.title('Average Tomato Yield by Year in China and India')
plt.show()
```
